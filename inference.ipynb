```python
from transformers import AutoTokenizer, AutoModelForCausalLM
from peft import PeftModel, PeftConfig
```

```python
base_model = AutoModelForCausalLM.from_pretrained("sarvamai/sarvam-1")
```

```python
config = PeftConfig.from_pretrained("finetuned-hindi-model-8k")
base = AutoModelForCausalLM.from_pretrained(config.base_model_name_or_path)
model = PeftModel.from_pretrained(base, "finetuned-hindi-model-8k")
tokenizer = AutoTokenizer.from_pretrained(config.base_model_name_or_path)
```

```python
tokenizer.pad_token = tokenizer.eos_token
```

```python
input_text = "डुकन ने बंदर खोली"
```

```python
input = f"Correct grammar for: {input_text}"
```

```python
tokenized_inputs = tokenizer(input, return_tensors="pt")
```

```python
outputs = model.generate(**tokenized_inputs, max_new_tokens=5)
result = tokenizer.decode(outputs[0])
print(result)
```
